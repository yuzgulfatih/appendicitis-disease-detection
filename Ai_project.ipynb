# **Simulated Annealing**

```python
class SimulatedAnnealing:
    def __init__(self, temp, alpha, stopping_temp, stopping_iter, model, data):
        ''' animate the solution over time

            Parameters
            ----------
            coords: array_like
                list of coordinates
            temp: float
                initial temperature
            alpha: float
                rate at which temp decreases
            stopping_temp: float
                temerature at which annealing process terminates
            stopping_iter: int
                interation at which annealing process terminates

        '''
        self.data = data
        self.model = model
        self.dimension = len(self.data.columns)-1
        self.temp = temp
        self.alpha = alpha
        self.stopping_temp = stopping_temp
        self.stopping_iter = stopping_iter
        self.iteration = 1

        self.curr_solution = np.random.rand(self.dimension)>0.5
        self.best_solution = self.curr_solution

        self.solution_history = [self.curr_solution]

        self.obj_value = self.objective_function(self.curr_solution)
        self.initial_weight = self.obj_value
        self.min_weight = self.obj_value

        self.weight_list = [-self.obj_value]

        print('First accuracy: ', -self.obj_value)

    def objective_function(self, sol):
      features=np.where(sol)[0]
      train_size = int(0.8*self.data.shape[0])
      train_data = self.data.iloc[0:train_size]
      test_data = self.data.iloc[train_size:]
      X_train = train_data.iloc[:,features].values
      Y_train = train_data.iloc[:,-1].values

      X_test = test_data.iloc[:,features].values
      Y_test = test_data.iloc[:,-1].values

      self.model.fit(X_train, Y_train)



      predicts = self.model.predict(X_test)

      acc = accuracy_score(Y_test, predicts)

      return -acc

    def acceptance_probability(self, candidate_weight):
        '''
        Acceptance probability as described in:
        https://stackoverflow.com/questions/19757551/basics-of-simulated-annealing-in-python
        '''
        return math.exp(-abs(candidate_weight - self.obj_value) / self.temp)

    def accept(self, candidate):
        '''
        Accept with probability 1 if candidate solution is better than
        current solution, else accept with probability equal to the
        acceptance_probability()
        '''
        candidate_weight = self.objective_function(candidate)
        if candidate_weight < self.obj_value:
            self.obj_value = candidate_weight
            self.curr_solution = candidate
            if candidate_weight < self.min_weight:
                self.min_weight = candidate_weight
                self.best_solution = candidate

        else:
            if random.random() < self.acceptance_probability(candidate_weight):
                self.obj_value = candidate_weight
                self.curr_solution = candidate

    def anneal(self):
        '''
        Annealing process with 2-opt
        described here: https://en.wikipedia.org/wiki/2-opt
        '''
        while self.temp >= self.stopping_temp and self.iteration < self.stopping_iter:
            candidate = list(self.curr_solution)
            random_index = np.random.randint(0,self.dimension)
            candidate[random_index] = ~ candidate[random_index]

            self.accept(candidate)
            self.temp *= self.alpha
            self.iteration += 1
            self.weight_list.append(-self.obj_value)
            self.solution_history.append(self.curr_solution)

        print('Best Accuracy: ', -self.min_weight)
        print('Improvement: ', -round((self.initial_weight - self.min_weight) / (self.initial_weight), 4) * 100, '%')
```
